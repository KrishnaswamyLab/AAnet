```python
import AAnet
import network
import numpy as np
import AAtools
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
import tensorflow as tf
import time
import pickle
from sklearn.decomposition import PCA
from PCHA import PCHA
from scipy.spatial import procrustes
import spams as sp
import NMF as nmf
import time
from sklearn.manifold import MDS
import phate
from sklearn.cluster import KMeans
from sklearn.decomposition import NMF
```

```python
# load mnist
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('/tmp/data/mnist', one_hot=False)
data = mnist.train.images
data = (data * 2) - 1 # norm for tanh
```

```python
# pick single digit
digit = 4
idx_digit = mnist.train.labels == digit
data = data[idx_digit,]
```

```python
# AAnet params
num_at = 3
noise_z_std = 0.03
z_dim = [256,64]
act_out = tf.nn.tanh
input_dim = data.shape[1]
```

```python
# network
enc_net = network.Encoder(num_at=num_at, z_dim=z_dim)
dec_net = network.Decoder(x_dim=input_dim, noise_z_std=noise_z_std, z_dim=z_dim, act_out=act_out)
model = AAnet.AAnet(enc_net, dec_net)
```

```python
# train
model.train(data, batch_size=128, num_batches=5000)
```

```python
ats_x = model.get_ats_x()
```

```python
# plot ATs
AAtools.plot_mnist(ats_x)
```

```python
# MDS
data_at = model.data2at(data)
embedding = MDS(n_components=2)
Y_mds_ats = embedding.fit_transform(ats_x)
Y_mds_data = data_at @ Y_mds_ats

fig, ax = plt.subplots(1, figsize=(4,3))
ax.scatter(Y_mds_data[:,0], Y_mds_data[:,1], s=.1, alpha=0.5)
ax.scatter(Y_mds_ats[:,0], Y_mds_ats[:,1], s=100, c='r')
for i in range(Y_mds_ats.shape[0]):
    ax.text(Y_mds_ats[i,0], Y_mds_ats[i,1], i+1, horizontalalignment='center', verticalalignment='center', 
             fontdict={'color': 'white','size':8,'weight':'bold'})
```

```python
# PCA on Z_data
pca = PCA(n_components=2)
Z_at = np.eye(num_at)
samples_Z = model.data2at(data)
Y_pca = pca.fit_transform(samples_Z)
Y_pca_at = pca.transform(Z_at)

fig, ax = plt.subplots(1, figsize=(4,3))
ax.scatter(Y_pca[:,0], Y_pca[:,1], s=.1, alpha=0.5)
ax.scatter(Y_pca_at[:,0], Y_pca_at[:,1], s=100, c='r')
for i in range(Y_pca_at.shape[0]):
    ax.text(Y_pca_at[i,0], Y_pca_at[i,1], i+1, horizontalalignment='center', verticalalignment='center', 
             fontdict={'color': 'white','size':8,'weight':'bold'})
```
