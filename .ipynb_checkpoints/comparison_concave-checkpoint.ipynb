```python
import AAnet
import network
import numpy as np
import AAtools
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
import tensorflow as tf
import time
import pickle
from sklearn.decomposition import PCA
from PCHA import PCHA
from scipy.spatial import procrustes
import NMF as nmf
import spams as sp
from sklearn.cluster import KMeans
from sklearn.decomposition import NMF
```

```python
# params fake data
n = 2000
nat = 3
ndim = 100
```

```python
# params AAnet
num_at = nat
noise_z_std = 0.05
z_dim = [100,100,100]
#z_dim = [] # linear AA
#act_out = tf.nn.tanh
act_out = None # linear last layer
input_dim = ndim
```

```python
# network
enc_net = network.Encoder(num_at=num_at, z_dim=z_dim)
dec_net = network.Decoder(x_dim=input_dim, noise_z_std=noise_z_std, z_dim=z_dim, act_out=act_out)
model = AAnet.AAnet(enc_net, dec_net)
```

```python
# fake data
z = np.random.uniform(size=[n,nat])
z = np.log(z)
z = z / np.sum(z, axis=1, keepdims=True)
at = np.random.uniform(size=[nat,ndim])
```

```python
# init
exp_vec = np.linspace(1, 0.25, 10)
```

```python
# AAnet
i = 0
mse_vec_aanet = []
at_list_aanet = []
for exp_val in exp_vec:
    print('i = %d, exp_val = %f' % (i,exp_val))
    z_concave = z**exp_val
    data = z_concave @ at
    
    # AAnet
    model.train(data, batch_size=128, num_batches=2000)
    AAnet_at = model.get_ats_x()
    _,_,MSE_AAnet = procrustes(AAnet_at.T, at.T)
    at_list_aanet.append(AAnet_at)
    print('MSE AAnet = %f' % MSE_AAnet)
    mse_vec_aanet.append(MSE_AAnet)
    
    i = i + 1
    
pickle_out = open("comparison_aanet.pickle","wb")
pickle.dump((mse_vec_aanet,at_list_aanet), pickle_out)
pickle_out.close()
```

```python
#mse_vec_aanet, at_list_aanet = pickle.load( open( "comparison_aanet.pickle", "rb" ) )
```

```python
# PCHA - Morup et al.
i = 0
mse_vec_pcha = []
at_list_pcha = []
for exp_val in exp_vec:
    print('i = %d, exp_val = %f' % (i,exp_val))
    z_concave = z**exp_val
    data = z_concave @ at
    
    # PCHA
    XC, _, _, _, _ = PCHA(data.T, noc=nat)
    _,_,MSE_PCHA = procrustes(XC, at.T)
    print('MSE PCHA = %f' % MSE_PCHA)
    at_list_pcha.append(XC.T)
    mse_vec_pcha.append(MSE_PCHA)
    i = i + 1
    
pickle_out = open("comparison_pcha.pickle","wb")
pickle.dump((mse_vec_pcha,at_list_pcha), pickle_out)
pickle_out.close()
```

```python
#mse_vec_pcha, at_list_pcha = pickle.load( open( "comparison_pcha.pickle", "rb" ) )
```

```python
# NMF - Javadi et al.
i = 0
mse_vec_nmf = []
at_list_nmf = []
for exp_val in exp_vec:
    print('i = %d, exp_val = %f' % (i,exp_val))
    z_concave = z**exp_val
    data = z_concave @ at
    
    # NMF
    W, H, L, Err = nmf.acc_palm_nmf(data, r=nat, maxiter=25)
    H = H.T
    _,_,MSE_nmf = procrustes(H, at.T)
    at_list_nmf.append(H)
    print('MSE nmf = %f' % MSE_nmf)
    mse_vec_nmf.append(MSE_nmf)
    
    i = i + 1
    
pickle_out = open("comparison_nmf.pickle","wb")
pickle.dump((mse_vec_nmf,at_list_nmf), pickle_out)
pickle_out.close()
```

```python
#mse_vec_nmf, at_list_nmf = pickle.load( open( "comparison_nmf.pickle", "rb" ) )
```

```python
# SPAMS - Mairal et al.
i = 0
mse_vec_spams = []
at_list_spams = []
for exp_val in exp_vec:
    print('i = %d, exp_val = %f' % (i,exp_val))
    z_concave = z**exp_val
    data = z_concave @ at
    data = data[:200,:]
    
    # SPAMS
    [Z, A, B] = sp.archetypalAnalysis(data.T, p=nat, returnAB=True, numThreads=-1)
    _,_,MSE_spams = procrustes(Z, at.T)
    at_list_spams.append(Z)
    print('MSE spams = %f' % MSE_spams)
    mse_vec_spams.append(MSE_spams)
    
    i = i + 1
    
pickle_out = open("comparison_spams.pickle","wb")
pickle.dump((mse_vec_spams,at_list_spams), pickle_out)
pickle_out.close()
```

```python
#mse_vec_spams, at_list_spams = pickle.load( open( "comparison_spams.pickle", "rb" ) )
```

```python
# kmeans clustering
i = 0
mse_vec_kmeans = []
at_list_kmeans = []
for exp_val in exp_vec:
    print('i = %d, exp_val = %f' % (i,exp_val))
    z_concave = z**exp_val
    data = z_concave @ at
    
    # kmeans
    kmeans = KMeans(n_clusters=nat, random_state=0).fit(data)
    Z = kmeans.cluster_centers_.T
    _,_,MSE_kmeans = procrustes(Z, at.T)
    at_list_kmeans.append(Z)
    print('MSE kmeans = %f' % MSE_kmeans)
    mse_vec_kmeans.append(MSE_kmeans)
    
    i = i + 1
    
pickle_out = open("comparison_kmeans.pickle","wb")
pickle.dump((mse_vec_kmeans,at_list_kmeans), pickle_out)
pickle_out.close()
```

```python
#mse_vec_kmeans, at_list_kmeans = pickle.load( open( "comparison_kmeans.pickle", "rb" ) )
```

```python
# PCA
i = 0
mse_vec_pca = []
at_list_pca = []
pca = PCA(n_components=nat)
for exp_val in exp_vec:
    print('i = %d, exp_val = %f' % (i,exp_val))
    z_concave = z**exp_val
    data = z_concave @ at
    
    # PCA
    pca.fit_transform(data)
    Z = pca.components_.T
    _,_,MSE_pca = procrustes(Z, at.T)
    at_list_pca.append(Z)
    print('MSE pca = %f' % MSE_pca)
    mse_vec_pca.append(MSE_pca)
    
    i = i + 1
    
pickle_out = open("comparison_pca.pickle","wb")
pickle.dump((mse_vec_pca,at_list_pca), pickle_out)
pickle_out.close()
```

```python
#mse_vec_pca, at_list_pca = pickle.load( open( "comparison_pca.pickle", "rb" ) )
```

```python
# NNMF
i = 0
mse_vec_nnmf = []
at_list_nnmf = []
for exp_val in exp_vec:
    print('i = %d, exp_val = %f' % (i,exp_val))
    z_concave = z**exp_val
    data = z_concave @ at
    data = data + (np.abs(np.random.normal(size=data.shape)) / exp_val / 2)
    # nnmf
    nnmf = NMF(n_components=nat, init='random', random_state=0)
    nnmf.fit_transform(data)
    H = nnmf.components_.T
    _,_,MSE_nnmf = procrustes(H, at.T)
    at_list_nnmf.append(H)
    print('MSE nnmf = %f' % MSE_nnmf)
    mse_vec_nnmf.append(MSE_nnmf)
    
    i = i + 1
    
pickle_out = open("comparison_nnmf.pickle","wb")
pickle.dump((mse_vec_nnmf,at_list_nnmf), pickle_out)
pickle_out.close()
```

```python
# 2
exp_vec = np.linspace(1, 3, 10)
i = 0
mse_vec_aanet = []
mse_vec_pcha = []
at_list_aanet = []
at_list_pcha = []
for exp_val in exp_vec:
    print('i = %d, exp_val = %f' % (i,exp_val))
    z_concave = z**exp_val
    data = z_concave @ at
    # AAnet
    model.train(data, batch_size=128, num_batches=2000)
    AAnet_at = model.get_ats_x()
    _,_,MSE_AAnet = procrustes(AAnet_at.T, at.T)
    at_list_aanet.append(AAnet_at)
    print('MSE AAnet = %f' % MSE_AAnet)
    mse_vec_aanet.append(MSE_AAnet)
    # PCHA
    XC, _, _, _, _ = PCHA(data.T, noc=nat)
    _,_,MSE_PCHA = procrustes(XC, at.T)
    print('MSE PCHA = %f' % MSE_PCHA)
    at_list_pcha.append(XC.T)
    mse_vec_pcha.append(MSE_PCHA)
    i = i + 1
```
