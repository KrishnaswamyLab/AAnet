```python
import AAnet
import network
import numpy as np
import AAtools
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
import tensorflow as tf
import time
import pickle
from sklearn.decomposition import PCA
from PCHA import PCHA
from scipy.spatial import procrustes
import spams as sp
import NMF as nmf
import time
from sklearn.manifold import MDS
import phate
from sklearn.cluster import KMeans
from sklearn.decomposition import NMF
```

```python
data = pickle.load(open('/home/dv292/data/celeba/celeba_56_50k.pkl', 'rb'))
D = 56
data = np.reshape(data, [-1, D*D*3])
data = data / 255
data = (data * 2) - 1 # norm for tanh
```

```python
# AAnet params
num_at = 100
noise_z_std = 0.0
act_out = tf.nn.tanh
nfilt = 16
```

```python
# network
enc_net = network.CelebA56ConvEncoder(num_at=num_at, nfilt=nfilt)
dec_net = network.CelebA56ConvDecoder(noise_z_std=noise_z_std, act_out=act_out, nfilt=nfilt)
model = AAnet.AAnet(enc_net, dec_net, gamma_convex=0, gamma_nn=0)
```

```python
#idx = np.random.randint(data.shape[0], size=500)
model.train(data, batch_size=256, num_batches=5000)
```

```python
# see recon
idx2 = np.random.randint(data.shape[0], size=10)
data_sample = data[idx2,:]
AAtools.plot_celeba(data_sample, nc=10, D=56)
data_sample_recon = model.at2data(model.data2at(data_sample))
AAtools.plot_celeba(data_sample_recon, nc=10, D=56)
```

```python
ats_all = np.ndarray([7,num_at,data.shape[1]])
```

```python
data_curr = data[:,:] # debug

pca = PCA(n_components=100)
pc = pca.fit_transform(data_curr)
U = pca.components_
mu = pca.mean_
```

```python
# AAnet
num_at = 10
noise_z_std = 0.07
act_out = tf.nn.tanh
nfilt = 16
enc_net2 = network.CelebA56ConvEncoder(num_at=num_at, nfilt=nfilt)
dec_net2 = network.CelebA56ConvDecoder(noise_z_std=noise_z_std, act_out=act_out, nfilt=nfilt)
model2 = AAnet.AAnet(enc_net2, dec_net2)
model2.train(data_curr, batch_size=256, num_batches=10000)
AT_aanet = model2.get_ats_x()
ats_all[0,:,:] = AT_aanet
```

```python
# PCHA
XC, S, C, _, _ = PCHA(pc.T, noc=num_at)
AT_pcha = XC.T @ U + mu
ats_all[1,:,:] = AT_pcha
```

```python
# SPAMS - Mairal et al.
#idx = np.random.randint(data_curr.shape[0], size=1000) # debug
[Z, A, B] = sp.archetypalAnalysis(np.asfortranarray(pc.T), p=num_at, returnAB=True, numThreads=-1)
AT_mairal = Z.T @ U + mu
ats_all[2,:,:] = AT_mairal
```

```python
# Javadi
W, H, L, Err = nmf.acc_palm_nmf(pc, r=num_at, maxiter=25)
AT_javadi = H @ U + mu
ats_all[3,:,:] = AT_javadi
```

```python
# kmeans
kmeans = KMeans(n_clusters=num_at, random_state=0).fit(pc)
AT_kmeans = kmeans.cluster_centers_
AT_kmeans = AT_kmeans @ U + mu
ats_all[4,:,:] = AT_kmeans
```

```python
# PCA
pca = PCA(n_components=num_at)
pca.fit_transform(data_curr)
AT_pca = pca.components_
ats_all[5,:,:] = AT_pca * 30
```

```python
# NMF
nnmf = NMF(n_components=num_at, init='random', random_state=0)
nnmf.fit_transform(data_curr+1)
AT_nmf = nnmf.components_
ats_all[6,:,:] = AT_nmf-1
```

```python
Q = np.reshape(ats_all, [7*num_at,56*56*3])
AAtools.plot_celeba(Q, nc=num_at, D=56)
```
